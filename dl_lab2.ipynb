```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

X = np.array([
    [1, 2],
    [2, 1],
    [2, 4],
    [3, 5],
    [6, 4],
    [7, 2]])
y = np.array([0, 0, 0, 0, 1, 1])

model = Perceptron(max_iter=1000, eta0=0.1, tol=None)
model.fit(X, y)

print("Weights (w):", model.coef_)
print("Bias (b):", model.intercept_)
test_point = np.array([[5, 3]])
prediction = model.predict(test_point)
print("Prediction for [5,3]:", prediction)

plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr')
w1, w2 = model.coef_[0]
b = model.intercept_[0]
x_vals = np.linspace(0, 8, 100)
y_vals = -(w1*x_vals + b) / w2
plt.plot(x_vals, y_vals)
plt.scatter(test_point[:,0], test_point[:,1], c='gold', s=100, edgecolor='black')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Single Unit Perceptron Decision Boundary")
plt.show()

X_and = np.array([[0,0],[0,1],[1,0],[1,1]])
y_and = np.array([0,0,0,1])
model_and = Perceptron(max_iter=1000, eta0=0.1, tol=None)
model_and.fit(X_and, y_and)
print("\nAND Gate:")
print("Actual:", y_and)
print("Predicted:", model_and.predict(X_and))
print("Accuracy:", accuracy_score(y_and, model_and.predict(X_and)))

X_or = np.array([[0,0],[0,1],[1,0],[1,1]])
y_or = np.array([0,1,1,1])
model_or = Perceptron(max_iter=1000, eta0=0.1, tol=None)
model_or.fit(X_or, y_or)
print("\nOR Gate:")
print("Actual:", y_or)
print("Predicted:", model_or.predict(X_or))
print("Accuracy:", accuracy_score(y_or, model_or.predict(X_or)))

X_xor = np.array([[0,0],[0,1],[1,0],[1,1]])
y_xor = np.array([0,1,1,0])
model_xor = Perceptron(max_iter=1000, eta0=0.1, tol=None)
model_xor.fit(X_xor, y_xor)
print("\nXOR Gate:")
print("Actual:", y_xor)
print("Predicted:", model_xor.predict(X_xor))
print("Accuracy:", accuracy_score(y_xor, model_xor.predict(X_xor)))
```
